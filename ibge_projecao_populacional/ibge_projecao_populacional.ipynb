```python
import polars as pl
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
```

```python
df = pl.read_ods('projecoes_2024_tab1_idade_simples.ods')
```

```python
df_long = df.unpivot(
    on=[str(ano) for ano in range(2000, 2071)],
    index=['IDADE', 'SEXO', 'LOCAL'],
    variable_name='ANO',
    value_name='POPULACAO'
)
```

```python
df_long
```

```python
df_br_ambos_homens_mulheres = df_long.filter(
    pl.col('LOCAL') == 'Brasil'
).group_by(
    ['SEXO', 'ANO']
).agg(
    pl.col('POPULACAO').sum()
).with_columns(
    pl.col('ANO').cast(pl.Int16)
).sort(['ANO', 'SEXO'])
```

```python
df_br_ambos_homens_mulheres
```

```python
df_br_ambos = df_br_ambos_homens_mulheres.to_pandas().query('SEXO == "Ambos"').reset_index(drop=True)
```

```python
df_br_ambos
```

```python
df_br_ambos.dtypes
```

```python
df_br_homens_mulheres = df_br_ambos_homens_mulheres.to_pandas().query('SEXO != "Ambos"').reset_index(drop=True)
```

```python
df_br_homens_mulheres
```

```python
fig = plt.figure(figsize=(12, 8), dpi=300)

gs = GridSpec(2, 2, figure=fig)

ax0 = fig.add_subplot(gs[0, :])
ax1 = fig.add_subplot(gs[1, :])

ax0.plot(
    df_br_ambos['ANO'],
    df_br_ambos['POPULACAO'],
    marker='o'
)

cores = plt.get_cmap('Dark2').colors
marcadores = ['o', '>']

for (sexo, grupo), cor, marcador in zip(df_br_homens_mulheres.groupby('SEXO'), cores, marcadores):
    ax1.plot(
        grupo['ANO'],
        grupo['POPULACAO'],
        color=cor,
        marker=marcador,
        label=sexo
    )

anos = list(range(2000, 2071, 5))
titulos = ['Sexo: Ambos', 'Sexo: Homens e Mulheres']

for ax, titulo in zip([ax0, ax1], titulos):
    ax.set_xticks(anos)
    ax.tick_params('x', rotation=45)
    ax.set_ylabel('População Projetada')
    ax.set_xlabel('Ano')
    ax.grid(color='grey', linestyle='--', alpha=0.5)
    ax.spines[['top', 'right']].set_visible(False)
    ax.set_title(titulo, fontweight='bold')

ax1.legend(loc='best')

plt.tight_layout()
fig.savefig('projecao_populacao_ibge_ambos_homens_mulheres_2000_2070.png', dpi=300, bbox_inches='tight')
plt.show()
```
